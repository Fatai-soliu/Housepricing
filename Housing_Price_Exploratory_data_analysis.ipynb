## Multiple Linear Regression -  Predicting House Price

```python
#Remove rows with nan

df = df[~np.isnan(df).any(axis=1)]
```

```python
X_multi = df.drop("price", axis=1)

Y_multi = df["price"]
```

```python
#X_multi.head()
Y_multi.head()
```

```python
X_multi_cons = sn.add_constant(X_multi)
```

```python
X_multi_cons.head()
```

```python
X_multi_cons.info()
```

```python
pd.isna(df)
```

```python
import numpy as np

# Check for infinite values in your DataFrame (assuming df is your DataFrame)
has_infinite_values = np.isinf(df.values).any()

print(has_infinite_values)
```

```python
# Assuming X_multi_cons is a pandas DataFrame
nan_indices = X_multi_cons.isna()
if nan_indices.any().any():
    print("NaN values found at indices:")
    print(nan_indices[nan_indices].stack().index.tolist())
```

```python
# Check for positive and negative infinity
inf_indices = np.isinf(X_multi_cons)
if inf_indices.any().any():
    print("Infinite values found at indices:")
    print(np.where(inf_indices))
```

No Inf

```python
lm_multi = sn.OLS(Y_multi, X_multi_cons).fit()
```

```python
lm_multi.summary()
```

## Test and Training Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_multi, Y_multi, test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape, X_test.shape, Y_test.shape, Y_train.shape)
```

```python
lm_a = LinearRegression()
```

```python
Y_test_a = lm_a.predict(X_test)
```

```python
Y_train_a = lm_a.predict(X_train)
```
